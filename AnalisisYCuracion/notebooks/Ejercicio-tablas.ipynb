## Tarea nro 1 percentiles

```python
percentiles = ['P3', 'P5', 'P10', 'P25', 'P50', 'P75', 'P90', 'P95', 'P97']
filtrado_masc = tabla_1[tabla_1.Sex==1]
masc_plot = filtrado_masc.plot(x='Agemos', y=percentiles)
masc_plot.set_ylabel('Peso')
masc_plot.set_title('Hombres')

filtrado_fem = tabla_1[tabla_1.Sex==2]
fem_plot = filtrado_fem.plot(x='Agemos', y=percentiles)
fem_plot.set_ylabel('Peso')
fem_plot.set_title('Mujeres')
```
