```python
def get_normalized_variable(dataset, variable):
    """
    This method will be used to normalize a variable within a defined dataset
    """
    mean, std = stats.norm.fit(dataset[variable].dropna())
    return (dataset[variable].dropna() - mean) / std
```

Realizaremos Gráficos QQ para comparar los datos de 'pf_identity' en latinoamérica y todo le mundo con una distribución normal.

```python
def get_norm_dist(dataset, variable):
    """
    Thismethod will return a normal distribution for a variable within a
    defined dataset.
    """
    mean, std = stats.norm.fit(dataset[variable].dropna())
    return stats.norm(mean, std)
```

```python
def graph_normal_distribution_distplot(dataset, variable, dist_label, x_label):
    """
    This method will graph a distplot to compare a variable within a
    dataset's distribution with a normal distribution
    """
    norm_dist = get_norm_dist(dataset, variable)
    sample = norm_dist.rvs(1000)

    plt.figure(figsize=(10,6))
    seaborn.distplot(sample, label='Normal Distribution')
    seaborn.distplot(dataset[variable].dropna(), label=dist_label)
    plt.legend(bbox_to_anchor=(1, 1), loc=2, borderaxespad=0.)
    plt.xlabel(x_label)
    seaborn.despine()
```

```python
def graph_qq_plot(dataset, variable, percs, x_label, y_label):
    """
    This method will draw a qq plot.
    """
    norm_dist = get_norm_dist(dataset, variable)
    qn_sample = np.percentile(dataset[variable].dropna(), percs)
    qn_norm_dist = np.percentile(norm_dist.rvs(len(dataset[variable].dropna())),
                                 percs)

    plt.figure(figsize=(15,10))
    seaborn.regplot(x=qn_sample, y=qn_norm_dist)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    x = np.linspace(np.min((qn_sample.min(), qn_norm_dist.min())),
                    np.max((qn_sample.max(),qn_norm_dist.max())))
    plt.plot(x,x, color=RED, ls="--")
    seaborn.despine()
```

```python
graph_normal_distribution_distplot(dataset, 'pf_identity',
                                   'Personal Freedom Identity - World',
                                   'Personal Freedom Identity')

graph_normal_distribution_distplot(latam_dataset, 'pf_identity',
                                   'Personal Freedom Identity - LatAm',
                                   'Personal Freedom Identity')
```

```python
graph_qq_plot(dataset, 'pf_identity', percs,
              'Personal Freedom Percentiles - World',
              'Normal Distribution Percentiles')
graph_qq_plot(latam_dataset,
              'pf_identity', percs,
              'Personal Freedom Percentiles - LatAm',
              'Normal Distribution Percentiles')
```

Realizaremos Gráficos QQ para comparar los datos de 'hf_score' en latinoamérica y todo le mundo con una distribución normal.
